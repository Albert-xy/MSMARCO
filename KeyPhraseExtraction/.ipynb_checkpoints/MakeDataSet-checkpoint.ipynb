```python
import numpy as np
import re
files = ['100k.tsv']
```

```python
urls, overlap, p = {}, {}, []
for file in files:
    with open(file,'r') as f:
        for l in f:
            l = l.strip().split('\t')
            if l[14] not in urls:
                urls[l[14]] = 0
            urls[l[14]] += 1
for file in files:
    with open(file,'r') as f:
        for l in f:
            l = l.strip().split('\t')
            print(len(l))
            if urls[l[14]] > 1 and len(l) > 16:
                if l[14] not in overlap:
                    overlap[l[14]] = []
                if 'error' not in l[16].lower(): 
                    overlap[l[14]].append(set(re.sub(' +', ' ',l[16].strip()).split(' ')))
```

```python
p = []
for url in overlap:
    gold = overlap[url][0]
    for i in range(1,len(overlap[url])):
        p.append(len(gold - overlap[url][i])/len(gold))
```
